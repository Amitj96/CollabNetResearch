```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
for j in range(1992,2019):
    add = "C://Users//amitd//Desktop//FA//Authors//AuthorlistPub-"+str(j)+".csv"
    newname = "AuthorFrequencyPub"+str(j)+".csv"
    df=pd.read_csv(add)
    df=df.drop(["id"],axis=1)
    nwlist=[]
    for i in range(len(df)) : 
        for column in df:
            elem = df.loc[i, column]
            if pd.isna(elem):
                continue
            #if elem in newlist:
             #   continue
            else:
                nwlist.append(elem)
            #print(df.loc[i, column]) 
    
    d=Counter(nwlist)
    df1 = pd.DataFrame.from_dict(d, orient='index').reset_index()
    df1=df1.rename(columns={'index':'Author',0:'Frequency'})
    df1.to_csv(newname,index=False)
```
