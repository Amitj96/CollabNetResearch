# Variance Inflation Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

```python
variables = ['B_AvgFirst7Years', 'B_AvgMid7Years', 'B_AvgLast7Years',
       'C_AvgFirst7Years', 'C_AvgMid7Years', 'C_AvgLast7Years',
       'D_AvgFirst7Years', 'D_AvgMid7Years', 'D_AvgLast7Years',
       'E_AvgFirst7Years', 'E_AvgMid7Years', 'E_AvgLast7Years']
```

```python
X = add_constant(dataset[variables])
```

```python
pd.Series([variance_inflation_factor(X.values, i) 
               for i in range(X.shape[1])], 
              index=X.columns)
```
